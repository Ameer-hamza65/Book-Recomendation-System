```python
ratings=pd.read_csv('BX-Book-Ratings.csv',sep=';',error_bad_lines=False,encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'},inplace=True)
```

```python
ratings.head()
```

```python
print('Books data: ',books.shape)
print('Users data: ',users.shape)
print("Ratings data: ",ratings.shape)
```

```python
x=ratings['user_id'].value_counts()>200
```

```python
x[x].shape
```

```python
y=x[x].index
```

```python
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings
```

```python
book_ratings=ratings.merge(books,on='ISBN')
```

```python
book_ratings
```

```python
number_rating=book_ratings.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'no_of_ratings'},inplace=True)
```

```python
number_rating
```

```python
final_rating=book_ratings.merge(number_rating,on='title')
```

```python
final_rating.head()
```

```python
final_rating=final_rating[final_rating['no_of_ratings']>=50]
```

```python
final_rating
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.head()
```

```python
final_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
final_pivot
```

```python
final_pivot.fillna(0,inplace=True)
```

```python
final_pivot
```

```python
#there are too many zeros in this dataset that will affect the recomendation so this csr_matrix will make algoritham 
#to not consider these zeros

from scipy.sparse import csr_matrix
book_sparse=csr_matrix(final_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances,suggestions=model.kneighbors(final_pivot.iloc[540,:].values.reshape(1,-1),n_neighbors=10)
```

```python
suggestions
```

```python
distances
```

```python
for i in range(len(suggestions)):
    print(final_pivot.index[suggestions[i]])
```

```python
final_pivot.index[54]
```

```python
def recomend_books(book_name):
    book_id=np.where(final_pivot.index==book_name)[0][0]
    distances,suggestions=model.kneighbors(final_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=10)
    
    for i in range(len(suggestions)):
        print(final_pivot.index[suggestions[i]])
```

```python
recomend_books('Exclusive')
```
